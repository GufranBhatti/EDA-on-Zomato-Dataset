```python
df_country = pd.read_excel(r'C:/Users/gufra/Downloads/Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
#How to check datatypes
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
#pie chart- Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: Maximum records or transactions are from India. After that USA and then United Kingdom

```python
final_df.columns
```

```python
#comparing three columns(features)
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
#Convert groupby object into dataframe.
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

Observations:
1. When rating is between 4.5 to 4.9 ---> Excellent
2. When rating is between 4.0 to 4.5 ---> Very Good
3. When rating is between 3.5 to 4.0 ---> Good
4. When rating is between 2.5 to 3.5 ---> Average
5. When rating is between 1.8 to 2.5 ---> poor
6. Many people did not rate.

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data = ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data = ratings,palette = ['blue','red','orange','yellow','green','green'])
```

Observations:
1. Not rated count is very high
2. Maximum ratings is between 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
